## Exploratory Analysis and Visualization

  <p style='text-align:justify;'> Data visualization is the representation of data through use of common graphics, such as charts, plots, infographics, and even animations. These visual displays of information communicate complex data relationships and data-driven insights in a way that is easy to understand.</p>
  <p style='text-align:justify;'>
Let's install and import "seaborn" library as "sns" and "matplotlib.pyplot" module as "plt" to perform some visualization operations on our data frame to understand the distribution and relationships of attributes.</p>


Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
import matplotlib.pyplot as plt

# Get value counts for car makes
car_make_counts = df['Car Make'].value_counts()

# Create a pie chart
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(car_make_counts, labels=car_make_counts.index, autopct='%1.1f%%', startangle=90)
ax.axis('equal')
ax.set_title('Distribution of Car Makes')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Get value counts for car makes
car_make_counts = df['Car Model'].value_counts()

# Create a pie chart
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(car_make_counts, labels=car_make_counts.index, autopct='%1.1f%%', startangle=90)
ax.axis('equal')
ax.set_title('Distribution of Car Models')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Group the data by year and calculate the total sales for each year
yearly_sales = df.groupby('Car Year')['Sale Price'].sum()

# Calculate the percentage of sales for each year
yearly_sales_perc = yearly_sales / yearly_sales.sum() * 100

# Create a line chart to visualize the yearly sales percentage
plt.plot(yearly_sales_perc.index, yearly_sales_perc.values)
plt.xlabel('Year')
plt.ylabel('Percentage of Sales')
plt.title('Yearly Sales Percentage')
plt.show()
```

  <p style='text-align:justify;'> The year 2015 having the highest sales and the year 2022 having the lowest sales.</p>

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

  <p style='text-align:justify;'> A correlation matrix heatmap is created using the seaborn (sns) and matplotlib (plt) libraries.</p>
  <p style='text-align:justify;'> The resulting heatmap provides a visual representation of the correlations between different numeric columns in the DataFrame. Positive correlations are indicated by warmer colors, negative correlations by cooler colors, and stronger correlations by darker shades. The annotated values on the heatmap cells give the precise correlation coefficients, indicating the strength and direction of the relationships between variables.
  
By analyzing the heatmap, patterns of correlation can be observed, helping to identify relationships between variables and understand the underlying patterns in the data.</p>

```python
plt.scatter(df['Car Make'], df['Sale Price'])
plt.xlabel('Car Make')
plt.ylabel('Sale Price')
plt.title('Sales Price v/s Car Make')
plt.show()
```

  <p style='text-align:justify;'>A scatter plot is created using the matplotlib (plt) library to visualize the relationship between the 'Car Make' column and the 'Sale Price' column of the DataFrame df.</p>
  <p style='text-align:justify;'>
The resulting scatter plot shows the distribution of car sales across different car makes, with the sale price indicated on the y-axis. Each data point represents a specific car sale, and its position on the plot reflects both the car make and the corresponding sale price. The scatter plot allows for a visual assessment of the relationship between car makes and their sale prices, providing insights into any potential patterns or trends.</p>

## Asking and Answering Questions


# Q1. What are the top car makes in terms of sales?

```python
top_car_makes = df['Car Make'].value_counts().head(10)
print(top_car_makes)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_car_makes.index, y=top_car_makes.values)
plt.xlabel('Car Make')
plt.ylabel('Count')
plt.title('Top 10 Car Makes')

# Rotate x-axis labels if needed
plt.xticks(rotation=45)

plt.show()
```

  <p style='text-align:justify;'> As we can see, the top car makers are Honda, followed by Chevrolet, Toyota, Ford, Nissan and all of them have sold at least 500000 cars.</p>

# Q2. How does the salesperson commission vary based on the car year?

```python
sns.boxplot(data=df, x='Car Year', y='Commission Earned')
plt.xticks(rotation=45)
plt.title('Range of Commission earned')
plt.show()
```

  <p style='text-align:justify;'> A box plot is created using the seaborn (sns) library to visualize the distribution of commission earned across different car years in the DataFrame df.
  
  The resulting box plot provides an overview of the distribution of commission earned for each car year. Each box represents the interquartile range (IQR) of the commission earned, with the horizontal line inside the box indicating the median value. The whiskers extending from the box show the range of the data, excluding outliers that are plotted as individual points.

The box plot allows for a visual comparison of commission earnings across different car years, highlighting any variations or trends. It can help identify car years associated with higher or lower commission earnings and potential outliers in the data.

As we can see, the commission earned throughout the years is always within the range of USD 2000 to USD 4000 from 2010 to 2022.</p>

#Q3: How much commission was earned with respect to the Sales Price?


```python
sns.scatterplot(data=df, x='Sale Price', y='Commission Earned')
plt.title('Commission earned v/s Sale Price')
plt.show()
```

<p style='text-align:justify;'>  As we can see from the scatter plot, the commission earned from the sales of cars increase as the price of the cars increase.

The commission earned is between USD 500 to USD 3000 when the car is priced between USD 10000 to USD 20000 and it gradually increases.

The commission earned when the price is USD 50000 can range anywhere between USD 2500 to USD 7000.</p>

# Q4: What is the distribution of Sales by month?

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
sales_by_month = df['Month'].value_counts().sort_index()
plt.figure(figsize=(8, 6))
sns.barplot(x=sales_by_month.index, y=sales_by_month.values)
plt.xlabel('Month')
plt.ylabel('Number of Sales')
plt.title('Distribution of Sales by Month')
plt.show()
```

  <p style='text-align:justify;'> The resulting bar plot visualizes the distribution of sales across different months. Each bar represents the number of sales made in a specific month, providing insights into the seasonal patterns or variations in sales throughout the year.
  
  The sales of cars was highest throughout the winter season (Nov-Dec). The sales otherwise stayed uniform throughout the year.</p>

# Q5: How to represent the Sales data using a Histogram?

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Sale Price', bins=20)
plt.xlabel('Sale Price')
plt.ylabel('Count')
plt.title('Distribution of Sale Prices')
plt.show()
```

<p style='text-align:justify;'> We can see the plot describes the number of cars with different price brackets.

The sales remained uniform throughout the entire price bracket, including the low-end cars and mid-range cars, even though we weould expect a decline in sales as the prices of the cars increases.</p>
